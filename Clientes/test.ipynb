```python
import pandas as pd
import geopandas as gpd
from shapely import wkb
import socket
import json
import time
import threading

from src.radom_json import radom_json_crudo
# Ruta al archivo .paquete que es un archivo Parquet
def enviar_json_thread(cliente_socket):
    try:
        path_json_rutas = 'Scripts/rutas.json'

        with open(path_json_rutas, 'r') as file:
            data = json.load(file)

        path_medellin = data['path_medellin']
        path_customers = data['path_customers']
        path_employees = data['path_employees']

        datos_json = radom_json_crudo(path_medellin, path_customers, path_employees)
        
        # Enviar el archivo JSON al cliente
        cliente_socket.sendall(json.dumps(datos_json).encode())
    except Exception as e:
        print("Error al enviar datos:", e)
    
    # Cerrar la conexión
    cliente_socket.close()

def servidor1():
    # Configuración del servidor 1
    HOST1 = '127.0.0.1'  # Dirección IP del servidor 1
    PUERTO1 = 65431      # Puerto del servidor 1
    MAX_ENVIOS = 5       # Número máximo de veces para enviar el JSON

    # Crear un socket TCP/IP para el servidor 1
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as servidor_socket1:
        servidor_socket1.bind((HOST1, PUERTO1))
        servidor_socket1.listen()

        print("Servidor 1 escuchando en {}:{}".format(HOST1, PUERTO1))
        contador = 0
        while contador < MAX_ENVIOS:
            conn, addr = servidor_socket1.accept()
            print("Conexión establecida con", addr)
            
            # Enviar el archivo JSON en un hilo separado
            thread_envio = threading.Thread(target=enviar_json_thread, args=(conn,))
            thread_envio.start()
            
            contador += 1
            print("Envío de JSON #{}".format(contador))
            time.sleep(3)  # Esperar 3 segundos antes de enviar nuevamente

def servidor2():
    # Configuración del servidor 2
    HOST2 = '127.0.0.1'  # Dirección IP del servidor 2
    PUERTO2 = 65432      # Puerto del servidor 2

    # Crear un socket TCP/IP para el servidor 2
    with socket.socket(socket.AF_INET, socket.SOCK_STREAM) as servidor_socket2:
        servidor_socket2.bind((HOST2, PUERTO2))
        servidor_socket2.listen()

        print("Servidor 2 escuchando en {}:{}".format(HOST2, PUERTO2))
        while True:
            conn, addr = servidor_socket2.accept()
            print("Conexión establecida con", addr)
            
            # Recibir el archivo JSON
            datos_recibidos = conn.recv(1024)

            # Verificar si se recibieron datos
            if datos_recibidos:
                # Decodificar los datos y cargar el JSON
                datos_json = json.loads(datos_recibidos.decode())
                print("Datos JSON recibidos:", datos_json)
            conn.close()  # Cerrar la conexión

# Crear y ejecutar los hilos para los servidores
thread_servidor1 = threading.Thread(target=servidor1)
thread_servidor2 = threading.Thread(target=servidor2)

thread_servidor1.start()
thread_servidor2.start()

# Esperar a que los hilos terminen
thread_servidor1.join()
thread_servidor2.join()
```
